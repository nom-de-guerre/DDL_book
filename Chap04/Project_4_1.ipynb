```python
def sigmoid (x):
    return 1/(1+exp(-x))

sigmoid_v = np.vectorize (sigmoid)
```

```python
def Softmax (logits):
    
    denom = 0
    for i in range (0, len (logits)):
        logits[i] = exp (logits[i])
        denom += logits[i]
        
    for i in range (0, len (logits)):
        logits[i] /= denom
        
    return logits
```

```python
def ClassifyIris (Weights, x):
    
    z = x
    Nlayers = len (Weights)
    
    for W in Weights:
    
        Nlayers -= 1
        
        z_ = W[:, 1:]@z
        z_ += W[:, 0:1]
        
        if Nlayers > 0:
            z = sigmoid_v (z_)
        else:
            z = Softmax (z_)

    return z
```
